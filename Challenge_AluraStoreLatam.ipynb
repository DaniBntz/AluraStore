#1. Análisis de facturación



```python
#Como primera medida, iniciamos importando las librerias.
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Cargo los archivos csv de las tiendas

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

#Asigno el valor correspondiente a cada tienda.

tienda1 = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)
```

```python
#Analizo las categorías dentro de cada archivo.
tienda.head()
```

```python

#lo siguiente es crear una columna a cada archivo en la que asigno el nombre de cada tienda
#así, al concatenar o unir los archivos podrán distinguirse

tienda1['tienda'] = 'Tienda 1'
tienda2['tienda'] = 'Tienda 2'
tienda3['tienda'] = 'Tienda 3'
tienda4['tienda'] = 'Tienda 4'

# Luego, procedo a calcular la facturación de cada tienda.

fact_tienda1 = tienda1['Precio'].sum()
fact_tienda2 = tienda2['Precio'].sum()
fact_tienda3 = tienda3['Precio'].sum()
fact_tienda4 = tienda4['Precio'].sum()

print(f'la facturación de la tienda 1 es: {fact_tienda1}')
print(f'la facturación de la tienda 2 es: {fact_tienda2}')
print(f'la facturación de la tienda 3 es: {fact_tienda3}')
print(f'la facturación de la tienda 4 es: {fact_tienda4}')


```

```python
#Ahora que tengo el resultado de la facturación por cada tienda
# crearé una tabla donde se vea la comparativa

import matplotlib.pyplot as plt

facturaciones = [fact_tienda1, fact_tienda2, fact_tienda3, fact_tienda4]
tiendas = ['tienda1', 'tienda 2', 'tienda 3', 'tienda 4']

plt.bar(tiendas, facturaciones, color='skyblue')
plt.title('Facturación total por tienda')

#Agrego los resultados de cada tienda
for i in range(len(tiendas)):
    plt.text(i, facturaciones[i] + 50000, f"${facturaciones[i]:,.0f}", ha='center', fontsize=9)

plt.show()

```

# 2. Ventas por categoría

```python
#Ahora paso a ver las categorías de venta por cada tienda.

cat_t1 = tienda1['Categoría del Producto'].value_counts()
print("Tienda 1 - Categorías más populares:")
print(cat_t1)

# Tienda 2
cat_t2 = tienda2['Categoría del Producto'].value_counts()
print("\nTienda 2 - Categorías más populares:")
print(cat_t2)

# Tienda 3
cat_t3 = tienda3['Categoría del Producto'].value_counts()
print("\nTienda 3 - Categorías más populares:")
print(cat_t3)

# Tienda 4
cat_t4 = tienda4['Categoría del Producto'].value_counts()
print("\nTienda 4 - Categorías más populares:")
print(cat_t4)
```

```python
# Ahora que tengo todas las categorías por tienda, escogeré las 5 más
# populares por cada una

#Tienda 1
cat_t1 = tienda1['Categoría del Producto'].value_counts().head(5)
print("Tienda 1 - Categorías más populares:")
print(cat_t1)

#Tienda 2
cat_t2 = tienda2['Categoría del Producto'].value_counts().head(5)
print("Tienda 2 - Categorías más populares:")
print(cat_t2)

#Tienda 3
cat_t3 = tienda3['Categoría del Producto'].value_counts().head(5)
print("Tienda 3 - Categorías más populares:")
print(cat_t3)

#Tienda 4
cat_t4 = tienda4['Categoría del Producto'].value_counts().head(5)
print("Tienda 4 - Categorías más populares:")
print(cat_t4)
```

```python
#Ahora, imprimo el gráfico correspondiente a las categorías más populares por tiendas

#creo los subplots, 2 gráficos arriba y 2 gráficos abajo

fig, axs = plt.subplots(2, 2, figsize=(14, 10))

#defino los valores de las tablas:

#Establezco los parámetos de la barra para Tienda 1
barra1 = axs[0, 0].bar(cat_t1.index, cat_t1.values, color='skyblue')
axs[0, 0].set_title('Tienda 1')
axs[0, 0].bar_label(barra1, padding=3)
axs[0, 0].tick_params(axis='x', rotation=45)

#Establezco los parámetos de la barra para Tienda 1
bars2 = axs[0, 1].bar(cat_t2.index, cat_t2.values, color='salmon')
axs[0, 1].set_title('Tienda 2')
axs[0, 1].bar_label(bars2, padding=3)
axs[0, 1].tick_params(axis='x', rotation=45)

#Establezco los parámetos de la barra para Tienda 1
bars3 = axs[1, 0].bar(cat_t3.index, cat_t3.values, color='lightgreen')
axs[1, 0].set_title('Tienda 3')
axs[1, 0].bar_label(bars3, padding=3)
axs[1, 0].tick_params(axis='x', rotation=45)

#Establezco los parámetos de la barra para Tienda 1
bars4 = axs[1, 1].bar(cat_t4.index, cat_t4.values, color='plum')
axs[1, 1].set_title('Tienda 4')
axs[1, 1].bar_label(bars4, padding=3)
axs[1, 1].tick_params(axis='x', rotation=45)


for ax in axs.flat:
    ax.set_ylabel('Cantidad de productos vendidos')

plt.tight_layout()
plt.suptitle('Top 5 Categorías Más Populares por Tienda', fontsize=16, y=1.03)
plt.show()
```
