#1. Análisis de facturación



```python
#Como primera medida, iniciamos importando las librerias.
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Cargo los archivos csv de las tiendas

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

#Asigno el valor correspondiente a cada tienda.

tienda1 = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)
```

```python
#Analizo las categorías dentro de cada archivo.
tienda.head()
```

```python

#lo siguiente es crear una columna a cada archivo en la que asigno el nombre de cada tienda
#así, al concatenar o unir los archivos podrán distinguirse

tienda1['tienda'] = 'Tienda 1'
tienda2['tienda'] = 'Tienda 2'
tienda3['tienda'] = 'Tienda 3'
tienda4['tienda'] = 'Tienda 4'

# Luego, procedo a calcular la facturación de cada tienda.

fact_tienda1 = tienda1['Precio'].sum()
fact_tienda2 = tienda2['Precio'].sum()
fact_tienda3 = tienda3['Precio'].sum()
fact_tienda4 = tienda4['Precio'].sum()

print(f'la facturación de la tienda 1 es: {fact_tienda1}')
print(f'la facturación de la tienda 2 es: {fact_tienda2}')
print(f'la facturación de la tienda 3 es: {fact_tienda3}')
print(f'la facturación de la tienda 4 es: {fact_tienda4}')


```

```python
#Ahora que tengo el resultado de la facturación por cada tienda
# crearé una tabla donde se vea la comparativa

import matplotlib.pyplot as plt

facturaciones = [fact_tienda1, fact_tienda2, fact_tienda3, fact_tienda4]
tiendas = ['tienda1', 'tienda 2', 'tienda 3', 'tienda 4']

plt.bar(tiendas, facturaciones, color='skyblue')
plt.title('Facturación total por tienda')

#Agrego los resultados de cada tienda
for i in range(len(tiendas)):
    plt.text(i, facturaciones[i] + 50000, f"${facturaciones[i]:,.0f}", ha='center', fontsize=9)

plt.show()

```

# 3. Calificación promedio de la tienda


```python
#Ahora, voy con el promedio de valoración de cada tienda.

import pandas as pd

#Lo que sigue es concatenar los Dataframes de cada tienda

df_tiendas = pd.concat([tienda1, tienda2, tienda3, tienda4], ignore_index = True)

#Ahora reviso el promedio de calificacion por tienda

calificacion_promedio = df_tiendas.groupby('tienda')['Calificación'].mean()
print(calificacion_promedio)

```

```python
#Ahora, de acuerdo al resultado anterior, paso a graficar
#los promedios de calificación por tienda.

import matplotlib.pyplot as plt

# Creo la gráfica
fig, ax = plt.subplots()
barras = ax.bar(calificacion_promedio.index, calificacion_promedio.values, color='mediumseagreen')

# Añado el título y las etiquetas
ax.set_title('Promedio de Calificación por Tienda')
ax.set_ylabel('Calificación Promedio')
ax.set_ylim(0, 5)

# Mostrar valores sobre de cada barra
for barra in barras:
    height = barra.get_height()
    ax.text(barra.get_x() + barra.get_width() / 2, height + 0.05, f'{height:.2f}',
            ha='center', va='bottom', fontsize=10)

# Mejorar visualización del eje X
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

# 4. Productos más y menos vendidos

```python
#Ahora reviso los productos más y menos vendidos por tienda

#Lo primero es filtrar por tienda, ya que anteriormente concatené los DataFrames de cada tienda en 1 solo

t1 = df_tiendas[df_tiendas['tienda'] == 'Tienda 1']
t2 = df_tiendas[df_tiendas['tienda'] == 'Tienda 2']
t3 = df_tiendas[df_tiendas['tienda'] == 'Tienda 3']
t4 = df_tiendas[df_tiendas['tienda'] == 'Tienda 4']

#Ahora, analizo los productos más y menps vendidos por tienda

#Tienda 1
mas_vendidos_t1 = t1['Producto'].value_counts().head(5)
menos_vendidos_t1 = t1['Producto'].value_counts().head(5)

#Tienda 2
mas_vendidos_t2 = t2['Producto'].value_counts().head(5)
menos_vendidos_t2 = t2['Producto'].value_counts().head(5)

#Tienda 3
mas_vendidos_t3 = t3['Producto'].value_counts().head(5)
menos_vendidos_t3 = t3['Producto'].value_counts().head(5)

#Tienda 4
mas_vendidos_t4 = t4['Producto'].value_counts().head(5)
menos_vendidos_t4 = t4['Producto'].value_counts().head(5)

#Mostrar resultados

print("Más vendidos Tienda 1:\n", mas_vendidos_t1)
print("Menos vendidos Tienda 1:\n", menos_vendidos_t1)

print("Más vendidos Tienda 2:\n", mas_vendidos_t2)
print("Menos vendidos Tienda 2:\n", menos_vendidos_t2)

print("Más vendidos Tienda 3:\n", mas_vendidos_t3)
print("Menos vendidos Tienda 3:\n", menos_vendidos_t3)

print("Más vendidos Tienda 4:\n", mas_vendidos_t4)
print("Menos vendidos Tienda 4:\n", menos_vendidos_t4)

```

```python
#Ahora que tengo los productos más y menos vendidos por tienda,
#realizo la visualización para que se entienda mejor

fig, axs = plt.subplots(4, 2, figsize=(16, 18))  # 4 tiendas, 2 columnas: más y menos vendidos

tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
datos_mas = [mas_vendidos_t1, mas_vendidos_t2, mas_vendidos_t3, mas_vendidos_t4]
datos_menos = [menos_vendidos_t1, menos_vendidos_t2, menos_vendidos_t3, menos_vendidos_t4]

for i in range(4):
    # Más vendidos
    axs[i, 0].barh(datos_mas[i].index[::-1], datos_mas[i].values[::-1], color='green')
    axs[i, 0].set_title(f'{tiendas[i]} - Más vendidos')
    axs[i, 0].set_xlabel('Cantidad')

    # Menos vendidos
    axs[i, 1].barh(datos_menos[i].index, datos_menos[i].values, color='red')
    axs[i, 1].set_title(f'{tiendas[i]} - Menos vendidos')
    axs[i, 1].set_xlabel('Cantidad')

plt.tight_layout()
plt.show()
```

```python
#Comparación más y menos vendidos por tienda en 1 sola gráfica.

#más vendidos
# Combino las Series en un solo DataFrame
productos_df = pd.concat(
    [mas_vendidos_t1, mas_vendidos_t2, mas_vendidos_t3, mas_vendidos_t4],
    axis=1
)
productos_df.columns = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']

# Rellenar con 0 para productos no presentes en todas las tiendas
productos_df.fillna(0, inplace=True)

# Crear gráfico agrupado
productos_df.plot(kind='bar', figsize=(16, 8))

plt.title('Productos más vendidos por tienda')
plt.ylabel('Cantidad de ventas')
plt.xlabel('Producto')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Tiendas')
plt.tight_layout()
plt.show()

#Menos vendidos
productos_df = pd.concat(
    [menos_vendidos_t1, menos_vendidos_t2, menos_vendidos_t3, menos_vendidos_t4],
    axis=1
)
productos_df.columns = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']

# Rellenar con 0 para productos no presentes en todas las tiendas
productos_df.fillna(0, inplace=True)

# Crear gráfico agrupado
productos_df.plot(kind='bar', figsize=(16, 8))

plt.title('Productos menos vendidos por tienda')
plt.ylabel('Cantidad de ventas')
plt.xlabel('Producto')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Tiendas')
plt.tight_layout()
plt.show()

```

#7.Resumen

```python
#Ahora voy a hacer un resumen donde analizaré los datos de la tienda menos rentable

#Facturación total por tienda

facturacion_total = df_tiendas.groupby('tienda')['Precio'].sum()

#Categoría más popular por tienda

categoria_popular = df_tiendas.groupby('tienda')['Categoría del Producto'].agg(lambda x: x.value_counts().idxmax())

#Promedio de calificación

promedio_calificacion = df_tiendas.groupby('tienda')['Calificación'].mean()


#productos más y menos vendidos - para este resumen seleccionaré 2 más y menos vendidos por tienda

productos_mas_vendidos = df_tiendas.groupby('tienda')['Producto'].value_counts().groupby('tienda').head(2)
productos_menos_vendidos = df_tiendas.groupby('tienda')['Producto'].value_counts().groupby('tienda').tail(2)

#Costo promedio de envíos

costo_envio_promedio = df_tiendas.groupby('tienda')['Costo de envío'].mean()


#Resumen

resumen = pd.DataFrame({
    'Facturación total': facturacion_total,
    'Calificación promedio': calificacion_promedio,
    'Envío promedio': costo_envio_promedio,
    'Categoría más popular': categoria_popular,
})
print(resumen)




```

```python
import matplotlib.pyplot as plt

# Datos
tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
facturacion = facturacion_total.values
calificaciones = promedio_calificacion.values
envios = costo_envio_promedio.values

# Como los valores para facturación son muy grandes, y no se representarían
# bien en una gráfica escalo todos los datos a una base común (0 a 100)
max_fact = max(facturacion)
facturacion_escalada = (facturacion / max_fact) * 100
envios_escalados = (envios / max(envios)) * 100
calificaciones_escaladas = (calificaciones / 5) * 100  # 5 es la calificación máxima

# Graficamos
plt.figure(figsize=(10, 6))

plt.plot(tiendas, facturacion_escalada, marker='o', label='Facturación (Escalada)', color='blue')
plt.plot(tiendas, calificaciones_escaladas, marker='o', label='Calificación Promedio (Escalada)', color='orange')
plt.plot(tiendas, envios_escalados, marker='o', label='Costo de Envío (Escalado)', color='green')

# Títulos y etiquetas
plt.title('Comparativa entre Tiendas (valores escalados)')
plt.xlabel('Tiendas')
plt.ylabel('Valores (Escala 0-100)')
plt.legend(loc='upper right')

# Texto adicional que explique la escala
plt.figtext(0.5, -0.05,
            "Nota: Todos los valores están escalados entre 0 y 100 para facilitar la comparación.\n"
            f"1 unidad en 'Facturación escalada' equivale aproximadamente a {round(max_fact / 100):,} pesos.",
            wrap=True, horizontalalignment='center', fontsize=9)

plt.grid(True)
plt.tight_layout()
plt.show()


```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Creamos el DataFrame resumen con valores reales
tabla_comparativa = pd.DataFrame({
    'Facturación Total': facturacion_total,
    'Calificación Promedio': promedio_calificacion,
    'Costo Envío Prom.': costo_envio_promedio
})

# Dibujamos la tabla
fig, ax = plt.subplots(figsize=(10, 2))
ax.axis('off')  # Oculta los ejes

tabla = ax.table(cellText=tabla_comparativa.round(2).values,
                 colLabels=tabla_comparativa.columns,
                 rowLabels=tabla_comparativa.index,
                 loc='center',
                 cellLoc='center')

tabla.auto_set_font_size(False)
tabla.set_fontsize(10)
tabla.scale(1.2, 1.5)  # Ancho, alto

plt.title("Resumen comparativo de tiendas", fontweight='bold')
plt.show()
```

```python
# Normalizar valores para poder compararlos
ranking = pd.DataFrame({
    'Puntuación Facturación Total': facturacion_total.rank(ascending=False),
    'Puntuación Calificación Promedio': promedio_calificacion.rank(ascending=False),
    'Puntuación Costo Envío Prom.': costo_envio_promedio.rank(ascending=True)
})

# Sumar las posiciones por tienda
ranking['Puntaje Total'] = ranking.sum(axis=1)

# Mostrar resultados
print("Ranking de tiendas (Más puntos = peor rendimiento general):")
print(ranking.sort_values(by='Puntaje Total', ascending=False))

# Determinar la tienda menos rentable
tienda_menos_rentable = ranking['Puntaje Total'].idxmax()
print(f"\n🔴 La tienda menos rentable es: {tienda_menos_rentable}")
```
