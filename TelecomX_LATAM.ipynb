# Extracci贸n

```python
import pandas as pd

dt = pd.read_json('/content/TelecomX_Data.json')
```

```python
dt.head()
```

```python
#primero voy a extraer la informaci贸n anidada en las columnas 'customer', 'phone', 'internet' y 'account'

customer = pd.json_normalize(dt['customer'])
phone = pd.json_normalize(dt['phone'])
internet = pd.json_normalize(dt['internet'])
account = pd.json_normalize(dt['account'])
```

```python
#ahora voy a unir todo en un solo DF para poder verlo mejor.

datos = pd.concat([dt, customer, phone, internet, account], axis=1)
```

```python
#ahora que ya tengo el DF con la informaci贸n completa, eliminar茅 las columnas
#customer, phone, internet, account, puesto que sobran en esta tabla ya que la
#informaci贸n ha sido extraida.

datos.drop(['customer', 'phone', 'internet', 'account'], axis=1, inplace=True)
#compruebo
datos.sample()
```

```python
datos.sample(5)
```

```python
datos['OnlineSecurity'].sample(20)
```

```python
#verifico la informaci贸n de mi DF

datos.info()
```

```python
#despues de revisar y analizar la informaci贸n, encuentro que la columna 'Charges.Total' es de tipo object
#deber铆a ser tipo float. voy a convertirla a tipo float64
#para esto primero necesito eliminar los caracteres que me impiden hacer la conversi贸n

datos['Charges.Total'] = datos['Charges.Total'].str.replace(',', '').replace('$', '').replace(' ', '')
```

```python
# Ahora, manejo expl铆citamente los strings vac铆os ('') resultantes de los reemplazos.
# Esto es crucial porque el paso anterior puede dejar strings vac铆os si la cadena original
# solo conten铆a los caracteres que se reemplazaron, o si ya estaba vac铆a.
# Reemplazo los strings vac铆os por None, que pandas manejar谩 como NaN (Not a Number)
# en una columna num茅rica. None es un valor que astype('float64') puede manejar.
datos['Charges.Total'] = datos['Charges.Total'].replace('', None)
```

```python
#ahora si voy a convertirla a tipo float64
datos['Charges.Total'] = datos['Charges.Total'].astype('float64')
```

```python
#compruebo
datos.info()
```

```python
# La columna 'SeniorCitizen' est谩 en formato 0 (No) y 1 (Yes).
# Verifico el tipo de dato actual antes de la conversi贸n
print(datos['SeniorCitizen'].dtype)
```

```python
# Realizo la conversi贸n a tipo booleano
datos['SeniorCitizen'] = datos['SeniorCitizen'].astype(bool)
```

```python
# Identifico que hay m谩s columnas de tipo object que deber铆an ser booleanas.
colum_bool = [
    'Partner',
    'Dependents',
    'PhoneService',
    'PaperlessBilling',
    'Churn'
]
```

```python
# Creo un diccionario de mapeo para convertir 'Yes' a True y 'No' a False.
mapeo1 = {'Yes': True, 'No': False}
```

```python
# Aplico el mapeo a las columnas seleccionadas
# Selecciono las columnas y luego uso .map() en ellas.
# Esto aplica la transformaci贸n a cada valor en las columnas especificadas.
for col in colum_bool:
    # Verifico si la columna existe en el DataFrame para evitar errores
    if col in datos.columns:
        # Utilizo .loc para asegurar que la asignaci贸n se haga correctamente en el DataFrame original
        datos.loc[:, col] = datos[col].map(mapeo1)
    else:
        print(f"La columna '{col}' no se encontr贸 en el DataFrame.")
```

```python
#ahora, hay otras columnas que tambi茅n deben ser transformadas a Booleano, pero
#dentro de algunas filas hay valores que impedir铆an la f谩cil conversi贸n, es por eso
#que estas columnas quiero cambiar primero algunos valores antes de transformarlas

colum_bool2 = [
    'MultipleLines',
    'InternetService',
    'OnlineSecurity',
    'OnlineBackup',
    'DeviceProtection',
    'TechSupport',
    'StreamingTV',
    'StreamingMovies'
              ]
```
