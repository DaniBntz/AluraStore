#ðŸ“Œ ExtracciÃ³n

```python
import pandas as pd

dt = pd.read_json('/content/TelecomX_Data.json')
```

```python
dt.head()
```

```python
#primero voy a extraer la informaciÃ³n anidada en las columnas 'customer', 'phone', 'internet' y 'account'

customer = pd.json_normalize(dt['customer'])
phone = pd.json_normalize(dt['phone'])
internet = pd.json_normalize(dt['internet'])
account = pd.json_normalize(dt['account'])
```

```python
#ahora voy a unir todo en un solo DF para poder verlo mejor.

datos = pd.concat([dt, customer, phone, internet, account], axis=1)
```

```python
#ahora que ya tengo el DF con la informaciÃ³n completa, eliminarÃ© las columnas
#customer, phone, internet, account, puesto que sobran en esta tabla ya que la
#informaciÃ³n ha sido extraida.

datos.drop(['customer', 'phone', 'internet', 'account'], axis=1, inplace=True)
#compruebo
datos.sample()
```

```python
datos.sample(5)
```

```python
datos['OnlineSecurity'].sample(20)
```

#ðŸ”§ TransformaciÃ³n

```python
#verifico la informaciÃ³n de mi DF

datos.info()
```

```python
datos['OnlineSecurity'].sample(5)
```

```python
#despues de revisar y analizar la informaciÃ³n, encuentro que la columna 'Charges.Total' es de tipo object
#deberÃ­a ser tipo float. voy a convertirla a tipo float64
#para esto primero necesito eliminar los caracteres que me impiden hacer la conversiÃ³n

datos['Charges.Total'] = datos['Charges.Total'].str.replace(',', '').replace('$', '').replace(' ', '')
```

```python
# Ahora, manejo explÃ­citamente los strings vacÃ­os ('') resultantes de los reemplazos.
# Esto es crucial porque el paso anterior puede dejar strings vacÃ­os si la cadena original
# solo contenÃ­a los caracteres que se reemplazaron, o si ya estaba vacÃ­a.
# Reemplazo los strings vacÃ­os por None, que pandas manejarÃ¡ como NaN (Not a Number)
# en una columna numÃ©rica. None es un valor que astype('float64') puede manejar.
datos['Charges.Total'] = datos['Charges.Total'].replace('', None)
```

```python
#ahora si voy a convertirla a tipo float64
datos['Charges.Total'] = datos['Charges.Total'].astype('float64')
```

```python
#compruebo
datos.info()
```

```python
# La columna 'SeniorCitizen' estÃ¡ en formato 0 (No) y 1 (Yes).
# Verifico el tipo de dato actual antes de la conversiÃ³n
print(datos['SeniorCitizen'].dtype)
```

```python
# Realizo la conversiÃ³n a tipo booleano
datos['SeniorCitizen'] = datos['SeniorCitizen'].astype(bool)
```

```python
#compruebo
datos.info()
```

```python
datos.sample(5)
```

```python
# Identifico que hay mÃ¡s columnas de tipo object que deberÃ­an ser booleanas.
colum_bool = [
    'Partner',
    'Dependents',
    'PhoneService',
    'PaperlessBilling',
    'Churn'
]
```

```python
# Creo un diccionario de mapeo para convertir 'Yes' a True y 'No' a False.
mapeo1 = {'Yes': True, 'No': False}
```

```python
# Aplico el mapeo a las columnas seleccionadas
# Selecciono las columnas y luego uso .map() en ellas.
# Esto aplica la transformaciÃ³n a cada valor en las columnas especificadas.
for col in colum_bool:
    # Verifico si la columna existe en el DataFrame para evitar errores
    if col in datos.columns:
        # Utilizo .loc para asegurar que la asignaciÃ³n se haga correctamente en el DataFrame original
        datos.loc[:, col] = datos[col].map(mapeo1)
    else:
        print(f"La columna '{col}' no se encontrÃ³ en el DataFrame.")
```

```python
#ahora si puedo transformar esas columnas en tipo bool
datos[colum_bool] = datos[colum_bool].astype('bool')
```

```python
#ahora, hay otras columnas que tambiÃ©n deben ser transformadas a Booleano, pero
#dentro de algunas filas hay valores que impedirÃ­an la fÃ¡cil conversiÃ³n, es por eso
#que estas columnas quiero cambiar primero algunos valores antes de transformarlas

colum_bool2 = [
    'MultipleLines',
    'InternetService',
    'OnlineSecurity',
    'OnlineBackup',
    'DeviceProtection',
    'TechSupport',
    'StreamingTV',
    'StreamingMovies'
              ]
```

```python
# Creo un diccionario de mapeo expandido para incluir el tercer valor.
mapeo2 = {
    'Yes': True,
    'No': False,
    'DSL': True,
    'Fiber optic': True,
    'No phone service': False,
    'No internet service': False # Incluyo este tambiÃ©n, es comÃºn en las columnas relacionadas con internet
}
```

```python
# Aplico el mapeo expandido a las columnas seleccionadas
for col in colum_bool2:
    # Verifico si la columna existe y es de tipo object antes de intentar mapear
    if col in datos.columns and datos[col].dtype == 'object':
         # Utilizo .loc para asegurar la asignaciÃ³n correcta
         datos.loc[:, col] = datos[col].map(mapeo2)
    elif col in datos.columns:
         print(f"La columna '{col}' no es de tipo object. No se aplicÃ³ el mapeo.")
    else:
        print(f"La columna '{col}' no se encontrÃ³ en el DataFrame.")
```

```python
#primero crearÃ© una lista con las columnas booleanas a transformar en int
colum_2int = [
    'Churn',
    'SeniorCitizen',
    'Partner',
    'Dependents',
    'PhoneService',
    'MultipleLines',
    'InternetService',
    'PaperlessBilling'
]
```

```python
#ahora voy a estandarizar los datos.
#los datos que tengo en booleano = True/False, los convertirÃ© en valores 1/0
# Convierto todas estas columnas (que ahora deberÃ­an ser 'bool') a tipo entero (int).
for col in colum_2int:
    # Verifico si la columna existe Y AHORA ES DE TIPO booleano antes de convertir a int
    if col in datos.columns and datos[col].dtype == bool:
        datos[col] = datos[col].astype(int)
    elif col in datos.columns:
         print(f"La columna '{col}' no es de tipo bool ({datos[col].dtype}). No se convirtiÃ³ a int.")
    else:
        print(f"La columna '{col}' no se encontrÃ³ en el DataFrame. Saltando conversiÃ³n a int.")
```
