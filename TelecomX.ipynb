#Extracción

```python
import pandas as pd
```

```python
dt = pd.read_json('/content/TelecomX_Data.json')
```

```python
dt
```

```python
#Primero Extraigola información anidada
customer = pd.json_normalize(dt['customer'])
phone = pd.json_normalize(dt['phone'])
internet = pd.json_normalize(dt['internet'])
account = pd.json_normalize(dt['account'])
```

```python
#ahora voy a unir todo en un solo DF para poder verlo mejor.

datos = pd.concat([dt, customer, phone, internet, account], axis=1)
```

```python
#compruebo
datos.sample(5)
```

```python
#Elimino las columnas que no me sirven.
datos.drop(['customer', 'phone', 'internet', 'account'], axis=1, inplace=True)

datos.sample(5)
```

#Transformación

```python
#reviso la información de las columnas
datos.info()
```

```python
#A primera vista noto que: Charges.Total, debe ser una columna tipo float.
#reviso la información que contiene la columna primero
datos['Charges.Total'].value_counts()
```

```python
import numpy as np
```

```python
#Ahora elimino los caracteres que me pueden impedir transformar a float
datos['Charges.Total'] = datos['Charges.Total'].str.replace(',', '').replace('$', '').replace(' ', '').replace('', None)
#Tranformo a float
datos['Charges.Total'] = datos['Charges.Total'].astype(np.float64)
#reviso
datos['Charges.Total'].info()
```

```python
cols_yes_no2 = ['MultipleLines','OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies']
```

```python
#hago una última comprobación de los datos.
for col in cols_yes_no2:
    print(f"{col}: {datos[col].unique()}")
```

```python
#Ahora, reemplazo los valores y cambio el tipo a booleano
mapeo = {
    'Yes': True,
    'No': False,
    'No phone service': False,
    'No internet service': False
}

for col in cols_yes_no2:
    datos[col] = datos[col].map(mapeo).astype(bool)

```

```python
datos[cols_yes_no2].info()
```

```python
datos.sample(5)
```

```python
#Ahora crearé la columna de cuentas diarias.

dias_mes = 30

datos['Cuentas_diarias'] = datos['Charges.Monthly'] / dias_mes

datos.head()
```

```python
# Ahora, noto que nos piden hacer una estandarización de los datos Yes/No o
# True/False a binario

#primero crearé una lista con las columnas booleanas a transformar en int
colum_2int = [
           'Churn', 'Partner', 'Dependents', 'PhoneService', 'PaperlessBilling',
           'MultipleLines', 'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
           'TechSupport', 'StreamingTV', 'StreamingMovies', 'SeniorCitizen'
             ]
```

```python
#ahora voy a estandarizar los datos.
#los datos que tengo en booleano = True/False, los convertiré en valores 1/0
# Convierto todas estas columnas (que ahora deberían ser 'bool') a tipo entero (int).
for col in colum_2int:
    # Verifico si la columna existe Y AHORA ES DE TIPO booleano antes de convertir a int
    if col in datos.columns and datos[col].dtype == bool:
        datos[col] = datos[col].astype(int)
    elif col in datos.columns:
         print(f"La columna '{col}' no es de tipo bool ({datos[col].dtype}). No se convirtió a int.")
    else:
        print(f"La columna '{col}' no se encontró en el DataFrame. Saltando conversión a int.")
```

```python
import matplotlib.pyplot as plt

churn_counts = datos['Churn'].value_counts()
labels = ['Se quedan', 'Se van']
colors = ['#66b3ff', '#ff6666']

plt.figure(figsize=(6, 6))
plt.pie(
    churn_counts,
    labels=labels,
    colors=colors,
    autopct='%1.1f%%',
    startangle=140,
    labeldistance=0.5,
    pctdistance=0.4
)
plt.title('Distribución de clientes (Churn)')
plt.axis('equal')
plt.show()
```
