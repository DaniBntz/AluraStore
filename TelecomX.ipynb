#Extracci√≥n

```python
import pandas as pd
```

```python
dt = pd.read_json('/content/TelecomX_Data.json')
```

```python
dt
```

```python
#Primero Extraigola informaci√≥n anidada
customer = pd.json_normalize(dt['customer'])
phone = pd.json_normalize(dt['phone'])
internet = pd.json_normalize(dt['internet'])
account = pd.json_normalize(dt['account'])
```

```python
#ahora voy a unir todo en un solo DF para poder verlo mejor.

datos = pd.concat([dt, customer, phone, internet, account], axis=1)
```

```python
#compruebo
datos.sample(5)
```

```python
#Elimino las columnas que no me sirven.
datos.drop(['customer', 'phone', 'internet', 'account'], axis=1, inplace=True)

datos.sample(5)
```

#Transformaci√≥n

```python
#reviso la informaci√≥n de las columnas
datos.info()
```

```python
#A primera vista noto que: Charges.Total, debe ser una columna tipo float.
#reviso la informaci√≥n que contiene la columna primero
datos['Charges.Total'].value_counts()
```

```python
datos['Charges.Total'].unique()
```

```python
import numpy as np
```

```python
#Ahora elimino los caracteres que me pueden impedir transformar a float
datos['Charges.Total'] = datos['Charges.Total'].str.replace(',', '').replace('$', '').replace(' ', '').replace('', None)
#Tranformo a float
datos['Charges.Total'] = datos['Charges.Total'].astype(np.float64)
#reviso
datos['Charges.Total'].info()
```

```python
datos['Charges.Total'].sample(10)
```

```python
datos.sample(5)
```

```python
datos.nunique()
```

```python
#como son muchos datos, revisar√© columna por columna

for col in datos.columns:
    print(f"\n--- {col} ---")
    print(datos[col].value_counts())
```

```python
#identifico que hay columnas que son tip object y deber√≠an ser booleanas
datos.info()
```

```python
#primero empiezo transformando la columna Senior citizen
datos['SeniorCitizen'] = datos['SeniorCitizen'].astype(bool)

datos['SeniorCitizen'].info()
```

```python
datos.sample(3)
```

```python
#ahora voy con las otras columnas tip object a booleano
#primero cambio los valores para poder pasarlos a booleano
cols_yes_no = ['Churn', 'Partner', 'Dependents', 'PhoneService', 'PaperlessBilling']
datos[cols_yes_no] = datos[cols_yes_no].replace({'Yes': True, 'No': False})
datos.sample(5)
```

```python
datos[cols_yes_no] = datos[cols_yes_no].astype(bool)
```

```python
cols_yes_no2 = ['MultipleLines','OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies']
```

```python
#hago una √∫ltima comprobaci√≥n de los datos.
for col in cols_yes_no2:
    print(f"{col}: {datos[col].unique()}")
```

```python
#Ahora, reemplazo los valores y cambio el tipo a booleano
mapeo = {
    'Yes': True,
    'No': False,
    'No phone service': False,
    'No internet service': False
}

for col in cols_yes_no2:
    datos[col] = datos[col].map(mapeo).astype(bool)

```
